# Sales data set details

```python
# load Sales data set    
p=r"E:\KUMAR\AIML\PROJECT2suppoting\DATASET\Sales.csv"
df_SAL=pd.read_csv(p,encoding='latin1')
df_SAL
```

```python
df_SAL.info()
```

```python
df_SAL.isnull().sum()
```

```python
# Convert dates to datetime 
df_SAL['Order Date'] = pd.to_datetime(df_SAL['Order Date'])

df_SAL['Delivery Date'] = pd.to_datetime(df_SAL['Delivery Date'])

# Calculate average days between order and delivery
mask = df_SAL['Delivery Date'].notna()
avg_days = round((df_SAL[mask]['Delivery Date'] - df_SAL[mask]['Order Date']).dt.days.mean())  
print(f"Average delivery days: {avg_days}")

# Fill missing delivery dates
df_SAL.loc[df_SAL['Delivery Date'].isna(), 'Delivery Date'] = df_SAL['Order Date'] + pd.Timedelta(days=int(avg_days))
```

```python
# Convert to SQL date format (YYYY-MM-DD)
df_SAL['Delivery Date'] = df_SAL['Delivery Date'].dt.strftime('%Y-%m-%d')
df_SAL['Order Date'] = df_SAL['Order Date'].dt.strftime('%Y-%m-%d')

# Verify results
print("\nSample of filled dates:")
print(df_SAL[['Order Date', 'Delivery Date']].head())
```

```python
print("\nNumber of missing values remaining:")
df_SAL.isnull().sum()
```

```python
# Rename some column

df_SAL = df_SAL.rename(columns={"Order Number": "Order_Number", "Line Item": "Line_Item", "Order Date": "Order_Date" , "Delivery Date": "Delivery_Date", "Currency Code": "Currency_Code"})

df_SAL   
```

# CREATE DATABASE 

```python
import mysql.connector as m
try:
    con = m.connect( host="localhost", user="root", password="admin" )
    cur=con.cursor()
    print("MySQL Database connection successful")
except Exception as err:
    print(f"Error: '{err}'")
```

```python
#create database
cur.execute("CREATE DATABASE IF NOT EXISTS DataSpark_project2")
cur.execute("use DataSpark_project2")
print('DB created')
```

# Create Table cus_details

```python
#create Customers table
cur.execute('''CREATE TABLE IF NOT EXISTS cus_details(
    CustomerKey int primary key, 
    Gender varchar(255),
    Name varchar(255), 
    City varchar(255), 
    State varchar(255),
    ZipCode varchar(255),         
    Country varchar(255), 
    Continent varchar(255), 
    Birthday varchar(255)
)''')
print('Table created successfully')
```

```python
df_cus['Birthday'] = df_cus['Birthday'].dt.strftime('%Y-%m-%d')  # Adjust format as needed
```

```python
df_cus = df_cus.apply(lambda x: x.str.encode('utf-8').str.decode('utf-8') if x.dtype == "object" else x)

```

```python
# Insert values to customer details table
ins_que = '''INSERT IGNORE INTO cus_details(CustomerKey, Gender, Name, City, State, ZipCode, Country, Continent, Birthday
) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s)'''

batch_size = 1000  # Adjust as needed

for i in range(0, len(df_cus), batch_size):
    batch_data = df_cus.iloc[i:i+batch_size].values.tolist()
    cur.executemany(ins_que, batch_data)
    con.commit()
print('successfully')
```
